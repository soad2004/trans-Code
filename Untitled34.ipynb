```python
import torch



X = torch.tensor([
    [1.0, 0.0, 1.0, 0.0],  # Token 1
    [0.0, 1.0, 0.0, 1.0],  # Token 2
    [1.0, 1.0, 0.0, 0.0],  # Token 3
    [0.0, 0.0, 1.0, 1.0],  # Token 4
    [1.0, 0.0, 0.0, 1.0],  # Token 5
    [0.0, 1.0, 1.0, 0.0],  # Token 6
    [1.0, 1.0, 1.0, 1.0],  # Token 7
])


d_model = X.shape[1]
d_k = 4


torch.manual_seed(0)

W_Q = torch.randn(d_model, d_k)
W_K = torch.randn(d_model, d_k)
W_V = torch.randn(d_model, d_k)


Q = X @ W_Q   # (7, 4)
K = X @ W_K   # (7, 4)
V = X @ W_V   # (7, 4)


scores = Q @ K.T   # (7, 7)


scores = scores / (d_k ** 0.5)

attention_weights = torch.softmax(scores, dim=1)
output = attention_weights @ V   # (7, 4)

print("Attention Scores:\n", scores)
print("\nAttention Weights:\n", attention_weights)
print("\nFinal Output:\n", output)
```
